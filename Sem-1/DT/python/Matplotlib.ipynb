```python
import matplotlib.pyplot as plt    # pyplot is sub package
```

LINE PLOTS (No Color specify)

```python
x = [0, 1, 2, 3, 4, 5, 6]           # x and y same dimension
y = [2,3,4,5,6,7,8 ]
plt.plot(x, y)
plt.show()
```

```python
plt.savefig("lineplot1.png", dpi=300)  # set the resolution
```

```python
x = [5, 2, 7]
y = [2, 16, 4]
plt.plot(x, y)
plt.title('Line Plot')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.show()
```

```python
x  = [1, 2, 3, 4, 5]
y1 = [50, 40, 70, 80, 20]
y2 = [20, 70, 50, 20, 60]
y3 = [70, 20, 60, 40, 60]
y4 = [80, 20, 20, 50, 60]
plt.plot(x, y1, 'green', label='day1', linewidth=3)
plt.plot(x, y2, 'b', label='day2', linewidth=3)
plt.plot(x, y3, 'r', label='day3', linewidth=3)
plt.plot(x, y4, 'm', label='day4', linewidth=3)
plt.title('Multiple Line Plot')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.legend()          
plt.show()
```

BAR PLOTS (Color specify)

```python
x = [60, 20, 30, 40, 50]
y = ["D", "E", "N", "P", "A"]
plt.bar(y, x, color = "b", width=0.5)
#plt.barh(y, x, color = "orange")
plt.show()
```

```python
x1 = [0.25, 1.25, 2.25, 3.25, 4.25]
y1 = [50, 40, 70, 80, 20]
plt.bar(x1, y1, label="BMW", color='g')
x2 = [.75, 1.75, 2.75, 3.75, 4.75]
y2 = [80, 20, 20, 50, 60]
plt.bar(x2, y2, label="Audi", color='m')
plt.xlabel('Days')
plt.ylabel('Distance')
plt.title('Two bar plots')
plt.legend()
plt.show()
```

```python
x1 = [0.25, 1.25, 2.25, 3.25, 4.25]
y1 = [50, 40, 70, 80, 20]
plt.bar(x1, y1, label="Enfield", color='y', width=2)

x2 = [0.26, 1.25, 2.25, 3.25, 4.25]
y2 = [80, 20, 20, 50, 60]
plt.bar(x2, y2, label="Honda", color='r', width=1)

x3 = [0.31, 1.5, 2.5, 3.5, 4.5]
y3 = [70, 20, 60, 40, 60]
plt.bar(x3, y3, label="Yamaha", color='m', width=.5)

x4 = [.75, 1.75, 2.75, 3.75, 4.75]
y4 = [80, 20, 20, 50, 60]
plt.bar(x4, y4, label="KTM", color='g', width=1.5)

plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance')

plt.title('Bikes details')

plt.show()
```

Scatter plot

```python
x = [5, 2, 9, 4, 7]     
y = [10, 5, 8, 4, 2]        
plt.scatter(y, x)      
plt.show()              
```

```python
x = [35, 20, 29, 40, 57]     # x-axis values 
y = [100, 50, 80, 40, 200]         # Y-axis values 

plt.scatter(x, y)         # Function to plot scatter 

plt.xlabel('Salary * 1000')
plt.ylabel('Age)')
plt.title('Age Vs Salary')

plt.show()               # function to show the plot 
```

```python
x1 = [1, 1.5, 2, 2.5, 3, 3.5, 3.6]
y1 = [7.5, 8, 8.5, 9, 9.5, 10, 10.5]

x2 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y2 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]

plt.scatter(x1, y1, label='High income low saving',color='r')
plt.scatter(x2, y2, label='Low income high savings', color='b')

plt.xlabel('Saving*100')
plt.ylabel('Income*1000')
plt.title('Scatter Plot')

plt.legend()

#plt.grid()
plt.show()      
```

Pie Plots

```python
grades = 'A', 'A-', 'B', 'B-', 'C', 'E'
gradeCount = [50, 60, 140, 160, 30, 20]

plt.pie(gradeCount, labels=grades, autopct='%1.2f%%') ##We use autopct to display the percent value 

plt.show()

```

```python
a = [3.5,6,2,1]
b= ['Enfield','Honda','Yahama','KTM']
plt.pie(a, labels=b, shadow= True, explode=(1,0,0.1,0.5), # explote spacing within the bikes
        autopct='%1.3f%%') # grey color shadow can be seen and 1.3 having three digits after decimal

plt.title('Bike Shares')

plt.show()
```
